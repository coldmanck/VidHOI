```python
import pandas as pd
import os, json
from tqdm import tqdm
import numpy as np

# generate tracklets
annot_dir = 'annotation/training'
tracklets = {}
for folder in tqdm(os.listdir(annot_dir)):
    for video_json in os.listdir(os.path.join(annot_dir, folder)):
        video_idx = folder + '/' + video_json.split('.')[0]
#         if video_idx != '0002/2739554651':
#             continue
        with open(os.path.join(annot_dir, folder, video_json), 'r') as f:
            annot = json.load(f)
#         import pdb; pdb.set_trace()
        
        tracklets_temp = {}
        for instance in annot['subject/objects']:
            if instance['category'] in ['adult', 'child', 'baby']:
                tracklets_temp[instance['tid']] = {}
                tracklets_temp[instance['tid']]['bbox'] = []
                tracklets_temp[instance['tid']]['frames'] = []
        for frame_id, trajectory in enumerate(annot['trajectories']):
            for instance in trajectory:
                if instance['tid'] in tracklets_temp:
                    coordinates = [
                        (instance['bbox']['xmin'] + instance['bbox']['xmax']) / 2, # center_x
                        (instance['bbox']['ymin'] + instance['bbox']['ymax']) / 2, # center_y
                        instance['bbox']['xmax'] - instance['bbox']['xmin'], # width
                        instance['bbox']['ymax'] - instance['bbox']['ymin'], # height
                    ]
                    tracklets_temp[instance['tid']]['bbox'].append(coordinates)
                    tracklets_temp[instance['tid']]['frames'].append(frame_id)
        
        # transform into ndarray
        for instance in tracklets_temp:
            tracklets_temp[instance]['bbox'] = np.array(tracklets_temp[instance]['bbox'], dtype=np.float64)
            tracklets_temp[instance]['frames'] = np.array(tracklets_temp[instance]['frames'])
        
        tracklets[video_idx] = tracklets_temp
```
