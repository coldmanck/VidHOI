```python
from tqdm import tqdm
# generate frame lists
annot_dir = 'annotation/validation'
trajectories = {}
frames = frames['results']
for folder in tqdm(os.listdir(annot_dir)):
    for video_json in os.listdir(os.path.join(annot_dir, folder)):
        with open(os.path.join(annot_dir, folder, video_json), 'r') as f:
            annot = json.load(f)
        video_idx = folder + '/' + video_json.split('.')[0]
        if video_idx not in trajectories:
            trajectories[video_idx] = [[] for _ in range(annot['frame_count'])]
        
        trajs = frames[video_idx]
        for tid, traj in enumerate(trajs):
            trajectory = traj['trajectory']
            for idx in range(traj['start_fid'], traj['end_fid']):
                bbox = trajectory[f'{idx:06d}']
                trajectories[video_idx][idx].append({
                    'tid': tid,
                    'bbox': {'xmin': bbox[0], 'ymin': bbox[1], 'xmax': bbox[2], 'ymax': bbox[3]},
                    'generated': 1,
                })
```

```python
with open('det_val_trajectories.json', 'w') as f:
    json.dump(trajectories, f)
```
